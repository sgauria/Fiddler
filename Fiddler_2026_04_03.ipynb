### This Week’s Fiddler

An alien—more specifically, an Eridian—named “Rocky” needs to pass a solid xenonite crystal to his human friend in a neighboring spaceship. The crystal is shaped like a regular tetrahedron, as shown below, and all its edges have length 1.

To safely transport the crystal, Rocky needs to create a long cylindrical tunnel between the spaceships, and then orient the tetrahedron so that it fits through the tunnel. It’s okay if the crystal fits snugly inside the tunnel—in this case, it can slide along without any friction.

What is the minimum possible radius for my tunnel so that the crystal will fit through it?

### This Week’s Extra Credit

Next, Rocky wants to transport a solid crystal shaped like a regular dodecahedron to his human friend. As before, each edge has length 1.

This time, the long tunnel between the space ships can be any right prismatic shape, not necessarily a cylinder. Once again, Rocky needs the crystal to fit through the tunnel, and it’s okay if that fit is snug.

What is the minimum possible cross-sectional area for the tunnel so that the crystal will fit through it?

# Fiddler Solution

Without proof, I'll claim that the smallest cross-sectional orientation is when one edge of the tetrahedron is aligned and touching the side of the cylinder parallel to the axis. The opposite edge will be perpedicular to the axis.

In this case, the cross-section will be a triangle with one side equalling 1 and the other 2 sides equalling the altitudes of the equilateral triangle = $\sqrt{3}/2$.

Here's a doodle on the tetrahedron image from the puzzle post showing which triangle I am talking about:



So, we need to find the circumradius of an isosceles triangle with base $1$ and sides $\sqrt{3}/2$, $\sqrt{3}/2$. This triangle will have altitude $\sqrt{(\sqrt{3}/2)^2 -(1/2)^2}$ = $\sqrt{3/4 - 1/4}$  = $1/\sqrt{2}$

Let the circumcenter by x units above the base. By setting the (squared) distances to the vertices to be equal, we get:

$(1/2)^2 + x^2 = ((1/\sqrt{2}) - x)^2$

$1/4 + x^2 = 1/2 -2x/\sqrt{2} + x^2$

$2x/\sqrt{2} = 1/4$

$x = \sqrt{2}/8 = 1/(4\sqrt{2})$

So, the circumradius is :

$R = (1/\sqrt{2}) -  1/(4\sqrt{2})$

$R = (1/\sqrt{2})( 1 -  1/4)$

$R = (1/\sqrt{2})(3/4)$

R = 0.707 ... * 0.75

**R = 0.53033008588991064330063327157864**

# Extra Credit Solution

Again without proof, I am going to claim that the smallest cross-sectional area is a hexagon that starts from the midpoint of the base of some pentagon, goes across to the apex of that pentagon, continues on the following edge, goes apex-to-base-midpoint in the following pentagon, and then goes base-midpoint-to-apex, edge, apex-to-base-midpoint, closing the loop.

My hand-wavy justification is that it makes sense to start with one pair of opposite faces to make the dodecahedron fit tightly. Then you pick some other face (rather than an edge for tightest packing), and that defines the axis direction of the prism. At that point, the other faces just follow.

I have taken the Dodecahedron projections from https://mathworld.wolfram.com/RegularDodecahedron.html and doodled versions of this hexagon in blue and yellow.




So, in short, we have a non-regular hexagon where 4 sides have length h (height of pentagon), and 2 sides have length 1. The 2 angles between the face-to-face transitions are the dihedral angle $d$. The 4 other angles are all identical, so we can calculate them based on the sum of angles of a hexagon, but it won't be needed.

It's easy to show that $h = cos(18\deg) + cos(54\deg)$, but https://en.wikipedia.org/wiki/Pentagon#Regular_pentagons has a closed-form expression: $h = (\sqrt{5 + 2\sqrt{5}})/2$



From https://mathworld.wolfram.com/RegularDodecahedron.html , the dihedral angle is $ d = cos^{-1}(-1/\sqrt{5}) $ = approx 116.565 degrees.



From this figure, we can calculate the area of the hexagon as the sum of the rectangular part + 4 times the right triangular parts.

$ A = 2h \times sin(d/2) \times 1 + 4 \times 1/2 \times h^2 \times sin(d/2)  cos(d/2)$

$ A = 2h \times sin(d/2) \times 1 + 2 \times h^2 \times sin(d/2)  cos(d/2)$

$ A = 2h \times sin(d/2) \times (1 + h \times cos(d/2))$

```python
from math import sqrt, sin, cos, radians, degrees, acos, isclose

h1 = sqrt(5 + 2 * sqrt(5)) / 2
h2 = cos(radians(54)) + cos(radians(18))
print (f"h = {h1}, h_alt = {h2}")
assert (isclose(h1, h2))
h = (h1+h2) / 2 # Make it closer to the actual value: https://www.wolframalpha.com/input?i=cos%2818deg%29+%2B+cos%2854deg%29

d = acos(-1 / sqrt(5))
assert(isclose(degrees(d), 116.565, abs_tol=1e-3))

d_div2 = d / 2

A_rect = 2 * h * sin(d_div2)

A_one_tri = h * h * sin(d_div2) * cos(d_div2) / 2

A_all_tri = 4 * A_one_tri

A = 2 * h * sin(d_div2) * (1 + h * cos(d_div2))

assert(isclose(A, A_rect + A_all_tri))

print (f"h = {h}, d = {degrees(d)}, d_div2 = {degrees(d_div2)}, A = {A}, A_rect = {A_rect}, A_all_tri = {A_all_tri}")
```

So, in conclusion, the cross-sectional area of the tunnel is about **4.7360679775**.
