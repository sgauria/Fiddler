Puzzle
----

https://xpuzzlecorner.substack.com/p/pirates-hate-this-one-trick

*Summary:*

What if the map had been folded n times? How many separate pieces of paper would there be if the map was cut perpendicular to the last fold?

General Principles
---

By visualizing what happens when we make a cut and then undo the last fold, we can infer some equivalences:

- $C$ cuts *perpendicular* to the last fold after $F$ folds is equivalent to $C$ cuts *parallel* to the last fold after $F-1$ folds.
- $C$ cuts *parallel* to the last fold after $F$ folds is equivalent to $2C$ cuts *perpedicular* to the last fold after $F-1$ folds.

Solution for n = 7
---

Using that, we can figure out tthe case of 7 folds followed by a perpendicular cut as follows.

1 cut perpendicular after 7 folds.
= 1 cut parallel after 6 folds.
= 2 cuts perpendicular after 5 folds.
= 2 cut parallel after 4 folds.
= 4 cuts perpendicular after 3 folds.
= 4 cut parallel after 2 folds.
= 8 cuts perpendicular after 1 folds.
= 8 cuts after 0 folds.

8 cuts parallel to each other will yield *9 pieces* of paper.

Solution for general case
----
Number of cuts increase in the 1x, 2x, 1x, 2x fashion until we reach 0 folds.

Number of cuts $C$ in general is:

$ C = 2^{floor(n/2)}$

And so the number of pieces $P$ is one more than that:

$ P = 1+ 2^{floor(n/2)}$


Verification
----
7 was a bit much, but I folded a paper $5$ times, cut it and verified that it was cut into $5 = 1 + 2^{floor(5/2)}$ pieces.
