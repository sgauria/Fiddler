### This Week’s Fiddler

This week’s puzzle was related to me by Jay Goldman:

I have two glasses, one containing precisely 12 fluid ounces of coffee, the other containing precisely 12 fluid ounces of tea.

I pour one fluid ounce from the coffee cup into the tea cup, and then thoroughly mix the contents. I then pour one fluid ounce from the (mostly) tea cup back into the coffee cup, and then thoroughly mix the contents.

Is there more coffee in the tea cup, or more tea in the coffee cup?

### This Week’s Extra Credit

I have two glasses that can hold a maximum volume of 24 fluid ounces. Initially, one glass contains precisely 12 fluid ounces of coffee, while the other contains precisely 12 fluid ounces of tea.

Your goal is to dilute the amount of coffee in the “coffee cup” by performing the following steps:

* Pour some volume of tea into the coffee cup.
* Thoroughly mix the contents.
* Pour that same volume out of the coffee cup (i.e., into the sink), so that precisely 12 fluid ounces of liquid remain.

After doing this as many times as you like, in the end, you will have 12 ounces of liquid in the coffee cup, some of which is coffee and some of which is tea. In fluid ounces, what is the least amount of coffee you can have in this cup?


# Fiddler Solution

The amount of coffee in the tea cup is the same as the amount of tea in the coffee cup.

This is because the initial and final amounts are the same, and all the tea and coffee is conserved.

More precisely, if *t* units of tea are transferred and *c* units of coffee are transferred, then:

*12 = 12 - t + c* // *because initial and final quantities in the tea cup are the same*

That simplifies to *t = c*. And similarly if you consider the coffee cup.

# Extra Credit Solution

For clarity, let's set 12 ounces = 1 unit, and do everything in these normalized units.

At some point, let $P$ be the proportion of coffee in the coffee cup, and let us do one step of dilution by adding $T$ units of tea to the "coffee cup", mixing it and then discarding $T$ units of the new mixture.

The new proportion of coffee will be:

$P_{1step} = \frac{P}{(1 + T)}$

And let us also consider what happens if we divide $T$ into $T_1$ and $T_2$, and dilute in 2 steps.

$P_{2steps} = \frac{P}{(1 + T_1)(1+T_2)}$

$P_{2steps} = \frac{P}{(1 + T_1+T_2 +T_1 T_2)}$

Now since, $T = T_1 + T_2$

$P_{2steps} = \frac{P}{(1 + T +T_1 T_2)}$

Comparing the 2 expressions, we can see that 

$P_{2steps} = \frac{P}{(1 + T +T_1 T_2)} < \frac{P}{(1 + T)} = P_{1step}$

$P_{2steps} < P_{1step}$

Since we did not make any assumptions on $P, T, T_1, T_2$, we can see that any dilution step can be improved by dividing it into 2 dilution steps, and then dividing further, and so on.

Taking this to the limit, the optimal scheme would be do the dilution in an infinite number of inifinitesimal steps.

Also we should use up all the tea for dilution, because there is no benefit to leaving some tea unused.

Putting that all together, we have:

$ P_{optimal} = \lim_{x \to 0}{(\frac{1}{1+x})^{1/x}}$

$ P_{optimal} = \lim_{x \to 0}{\frac{1}{(1+x)^{1/x}}}$

This is a well known (but not intuitive, at least for me) limit and the answer is 1/e. 

I am not going to try to explain how it is reached, except for linking to some sites that do that:
https://www.math-linux.com/mathematics/limits/article/proof-of-limit-of-lim-1-x-1-x-e-as-x-approaches-0  
https://www.mathdoubts.com/lim-1-plus-1-by-x-whole-power-x-as-x-approaches-infinity-proof/ 

$P_{optimal} = \frac{1}{e}$

In the original units, the minimum amount of coffee left the coffee cup would be 12/e ounces = 4.4145532940573078591462852419375 ounces.

We can do a fun plot to see how many steps are needed to approximate the optimal.

```python
num_steps = list(range(1,251))
num_steps.sort()

proportions = [(1/(1+(1/ns)))**ns for ns in num_steps]

import matplotlib.pyplot as plt 
from math import e
plt.plot(num_steps, proportions, color='violet')
plt.xlabel('Number of Dilution Steps')
plt.ylabel('Proportion of Coffee left')
plt.axhline(1/e, color='red', linestyle='--', label='1/e')
plt.title('Proportion of Coffee Left vs Number of Dilution Steps')
plt.legend()
plt.show()
```
