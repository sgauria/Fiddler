### This Week’s Fiddler

The Fiddler Basketball Association’s All-Star Game consists of two teams: “East” and “West.” Every year these two teams play a game, each with a 50 percent chance of winning that’s independent of the outcomes of previous years.

Many, many years into the future, you look at the most recent results of the All-Star Game. On average, what is the longest current winning streak that one of the teams is on? (Here, having won only the most recent game still counts as a “streak” of one game.)

### This Week’s Extra Credit

To spice up the All-Star Game, the commissioner of the FBA has decided that there will now be three teams competing in All-Star Games: “Stars,” “Stripes,” and “International.” Each year, two of the three teams play each other. If one year has Stars vs. Stripes, the next year has Stripes vs. International, the year after that has International vs. Stars, and then the cycle repeats with Stars vs. Stripes.

Many, many years after this new format has been adopted, you look at the most recent results of the All-Star game. On average, what is the longest current winning streak that one of the teams is on? (As before, having won one game counts as a “streak.” Also, note that the team with the longest winning streak might not be one of the two teams that played in the most recent All-Star Game.)

# Fiddler Solution

Let the max current streak length at some point be S.

After one more game, the streak will either be S+1 (if the same team wins) or 1 (if the other team wins).

In equilbrium, these should be the same.

i.e. S = 1/2 * (S+1) + 1/2 * 1

which gives us 

S = 2.

# Extra Credit Solution

In any set of 3 successive games, there will be at least 2 winners and 2 losers since no team plays all the games.

So, there must be at least one team that both loses and wins, and this team's streak wil be either 0 or 1.

At some point, let A be the team with the max current streak M, which must be at least 1 (and if 1, could be joint). 

Regarding B and C, either both are at 0, or one of them is at 1 and the other is 0.

After 3 more games, each team will have played 2 games, and the situation could be as follows:

| A v B | A v C | B v C | M' = New Max Current Streak | Comments |
|-|-|-|-|-|
| A | A | B | M+2 | A=M+2. B=1. C=0. |
| A | A | C | M+2 | A=M+2. B=0. C=1. |
| A | C | B | 1 | A=0. B=1. C=0. |
| A | C | C | 2 | A=0. B=0. C=2. |
| B | A | B | 2 | A=1. B=2. C=0.|
| B | A | C | 1 | A=1. B=0. C=1. |
| B | C | B | 2 | A=0. B=2. C=0. |
| B | C | C | 2 | A=0. B=0. C=2 |

So, we can say that

M' = 1/8 ( 2 * (M+1) + 2 * 1 + 4 * 2 )

M' = 1/8 ( 2M + 4 + 8)

M' = 2M/8 + 12/8

M' = M/4 + 3/2

In equilibrium, M' = M, so we can solve for M.

3M/4 = 3/2

M = 2

Whoa. It is the same.

That could make sense, since from any single team's perspective the situation is essentially the same, and the expected streak length for any particular team is the same.
