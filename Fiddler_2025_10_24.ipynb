**This Week’s Fiddler**

With the regular season over, there are two clear favorites for baseball’s American League Most Valuable Player (MVP) award according to [ESPN](https://www.espn.com/sports-betting/story/_/id/44476769/2025-mlb-odds-tracker-mvp-cy-young-rookie-year-espnbet):

    Aaron Judge of the New York Yankees, whose odds are -150.
    Cal Raleigh of the Seattle Mariners, whose odds are +110.

The “-150” odds mean that for every $150 you bet on Judge to win MVP, you’ll earn $100 if he actually wins. The “+110” odds mean that for every $100 you bet on Raleigh to win MVP, you’ll win $110 if he actually wins. Note the differing results (winning $100 vs. betting $100) that comes with the odds being negative vs. positive.

While these betting lines may be informed by an assessment of Judge’s and Raleigh’s real chances, they may also be informed by how much money people are betting on each player.

Suppose all bettors have wagered on either Judge or Raleigh with the odds above. Some fraction f of dollars wagered have been in favor of Judge, while 1−f has been wagered on Raleigh. For what fraction f will the oddsmaker earn the same amount of money, regardless of which player earns the MVP award?

**This Week’s Extra Credit**

Suppose there are two leading candidates, A and B, for MVP in the Fiddler Baseball League. There are two parts to this Extra Credit, so please read carefully!

Part 1:

The odds for A winning the award have been set to +100x, where x > 1.

Let f represent the fraction of dollars wagered in favor of A. For many values of f, the oddsmaker can set the odds for B so that they’ll make the same amount of money regardless of whether A or B wins the award. However, below a certain value of f, it’s impossible for the oddsmaker to do this.

What is this critical value of f? (Your answer should be in terms of x.)

Part 2:

Now, the odds for A winning the award have been set to -100y, where y >1. Again, for many values of f, the oddsmaker can set the odds for B so they’ll make the same amount whether A or B wins the award.

What is the critical value of f below which this isn’t possible? (Your answer should be in terms of y.)

# Fiddler Solution

Let M denote money out from the oddsmaker's point of view, i.e. how much money they would have to pay out as return of stake + winnings.

Let $M_A$ be the money out for player A (Aaron Judge) and $M_B$ be the money out for player B (Cal Raleigh). Let T be the total amount of money at stake, i.e. all the money that has been bet on either A or B.

The payout ratio for A (-150) is (100 + 150) / 150 = 250 / 150 = 5/3 = 1.667

The payout ratio for B (+110) is (100 + 110) / 100 = 210 / 100 = 21/10 = 2.1

$M_A = T * f * 5/3$

$M_B = T * (1-f) * 21/10$

In either case, the oddsmaker's earnings are T - M.

So, setting:

$T - M_A = T - M_B$

$M_A = M_B$

$T * f * 5/3 = T * (1-f) * 21/10$ 

$f * (5/3 + 21/10) = 21/10$ 

Multiplying throughout by 30.

$f * (50 + 63) = 63$

$f = 63 / 113 = 0.55752212389380530973451327433628$

Makes sense that f is greater than 0.5. If A wins, then the money out is smaller per stake but over a larger number of dollars wagered. And vice versa when B wins.

# Extra Credit Solution - Part 1

Odds for A are +100x

Payout ratio for A is (100 + 100x)/100 = 1 + x. Let's call this $P_A$.

And let $P_B$ be the payout ratio for B.

Following steps similar to the Fiddler, we'll get

$f * P_A = (1-f) * P_B$

$P_B = f * P_A / (1-f)$

$P_B = (f / (1-f)) * (1 + x)$

Now, this payout ratio $P_B$ must be something greater than 1, for betting on B to make sense. So, we can find the critical fraction f by setting this to 1.

$(f / (1-f)) * (1 + x) > 1$

$f(1+x) > 1-f$

$f(1+x) + f > 1$

$f(2+x) > 1$

$f > 1/(2+x)$

As a cross-check, let's consider the case of +110 odds from the Fiddler. x = 1.1. So f > 1/3.1 = 0.32.

i.e. If less than ~1/3 of the bets are on Cal (and these bets amplify money by a factor of 2.1), there is no way to make the same amount of money regardless of who wins. This makes sense if you consider the extreme case - e.g. only 1% of the money is on Cal. Then the payout when Cal wins will be pretty small even after the amplification factor. Since the money at stake on Aaron (99%) is much larger than the amplified amount (2.1%), so equalizing them will mean an nonsensical bet on Aaron.

# Extra Credit Solution - Part 2

Odds for A are -100y

Payout ratio for A is (100 + 100y)/100y = 1 + 1/y. Let's call this $P_A$.

And let $P_B$ be the payout ratio for B.

Following steps similar to the Fiddler, we'll get

$f * P_A = (1-f) * P_B$

$P_B = f * P_A / (1-f)$

$P_B = (f / (1-f)) * (1 + 1/y)$

Now, this payout ratio $P_B$ must be something greater than 1, for betting on B to make sense. So, we can find the critical fraction f by setting this to 1.

$(f / (1-f)) * (1 + 1/y) > 1$

$f(1+1/y) > 1-f$

$f(1+1/y) + f > 1$

$f(2+1/y) > 1$

$f > 1/(2+1/y)$

$f > y/(2y+1)$

This is similar to the other case, with x replaced by 1/y.

# Conclusion

Fiddler Answer: $f = 63 / 113 = 0.55752212389380530973451327433628$

Extra Credit Answer Part1 : $f = 1/(2+x)$

Extra Credit Answer Part1 : $f = 1/(2+(1/y)) = y/(2y+1)$
