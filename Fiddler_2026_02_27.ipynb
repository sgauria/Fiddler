### This Week’s Fiddler

In 2026, every day of the week is the first day of the month at least once:

* Monday is June 1.
* Tuesday is September 1 and December 1.
* Wednesday is April 1 and July 1.
* Thursday is January 1 and October 1.
* Friday is May 1.
* Saturday is August 1.
* Sunday is February 1, March 1, and November 1.

Is 2026 special in this regard? If so, when is the next year when one of the days of the week is not represented among the firsts of the month? Otherwise, if 2026 is not special in this regard, then why not?

### This Week’s Extra Credit

As we just noted, in 2026, all seven days of the week appear as the first of the month at least once. But you know, I decided that I don’t like that at all. Instead, I want as few days of the week as possible to appear as the first of the month in a given year.

To accomplish this, I have been granted the authority to change the number of days in each of that year’s 12 months, provided that there are still 365 or 366 days in the year and each month has at least 28 days and at most 31 days.

What are the fewest days of the week that can appear as the first of the month in such a calendar year? (And for fun, rather than for credit: How many such calendars can you design with this property?)

# Solution

Let's consider mapping the days of the week to the numbers 0 to 6 modulo 7, in some arbitrary rotated fashion. Then, assuming Jan 1 starts at 0, it is pretty straightforward to calculate the start days of all the other months. And since 0 could be mapped to any of the 7 days of the week, we'll have considered all possible calendars. We just need to also check leap years, which have a different structure than regular years.

```python
def distinct_days(calendar, verbose=False):
    days = []
    for i in range(len(calendar)):
        day = sum(calendar[:i]) % 7
        days.append(day)
    distinct_days = set(days)
    if verbose:
        print(f"{calendar} (months) -> {days} (start days) -> {distinct_days} (distinct start days)")
    return len(distinct_days)
```

```python
distinct_days([31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31], verbose=True)
```

```python
distinct_days([31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31], verbose=True)
```

So, we can see that any regular year or leap year will have all 7 days among the starts of the months.

In other words, there will never be a year (in the gregorian calendar) where some day of the week is not a first of any month.

# Extra Credit.

Since the number of cases is relatively few, we can check all possible calendars.

```python
from itertools import product
month_options = [28,29,30,31]

distinct_days_counts = {}
for i in range(0,8):
    distinct_days_counts[i] = 0

for calendar in product(month_options, repeat=12):
    if sum(calendar) != 365 and sum(calendar) != 366:
        continue
    dd = distinct_days(calendar)
    distinct_days_counts[dd] += 1
    if dd <= 3:
        distinct_days(calendar, verbose=True)

print("\nDistinct days counts:")
for dd in sorted(distinct_days_counts.keys()):
    print(f"{dd}: {distinct_days_counts[dd]}")   
```

So, we can see that the minimum number of distinct days of week that can appear as firsts of the months is 3 days.

There are 4 different calendars that can do this.
* (30, 31, 30, 30, 31, 30, 30, 31, 30, 30, 31, 31) (months) -> [0, 2, 5, 0, 2, 5, 0, 2, 5, 0, 2, 5] (start days) -> {0, 2, 5} (distinct start days)
* (31, 30, 30, 31, 30, 30, 31, 30, 30, 31, 30, 31) (months) -> [0, 3, 5, 0, 3, 5, 0, 3, 5, 0, 3, 5] (start days) -> {0, 3, 5} (distinct start days)
* *(31, 31, 29, 31, 31, 29, 31, 31, 29, 31, 31, **30**) (months) -> [0, 3, 6, 0, 3, 6, 0, 3, 6, 0, 3, 6] (start days) -> {0, 3, 6} (distinct start days)*
* *(31, 31, 29, 31, 31, 29, 31, 31, 29, 31, 31, **31**) (months) -> [0, 3, 6, 0, 3, 6, 0, 3, 6, 0, 3, 6] (start days) -> {0, 3, 6} (distinct start days)*

And the last two differ in just one month, making them ideal candidates for a regular year / leap year pair, featuring December(!) as the leap month with 30 or 31 days.

And it is also worth noting that if we could have 13 months, then we could have a calendar with all months 28 days except the last one, which would have 29 or 30 days, and then all months would start on the same day of the week!
